```python
# Create a copy of the crowdfunding_info_df DataFrame name campaign_df. 
campaign_df = crowdfunding_info_df.copy()
campaign_df.head()
```

```python
# Rename the blurb, launched_at, and deadline columns.
campaign_df = campaign_df.rename(columns={
    'blurb': 'description',
    'launched_at': 'launch_date',
    'deadline': 'end_date'
})

# Display the updated DataFrame
campaign_df.head()
```

```python
# Convert the goal and pledged columns to a `float` data type.
campaign_df['goal'] = campaign_df['goal'].astype(float)
campaign_df['pledged'] = campaign_df['pledged'].astype(float)

# Display the updated DataFrame
campaign_df.head()
```

```python
# Check the datatypes
print(campaign_df.dtypes)
```

```python
import pandas as pd
from datetime import datetime as dt

# Format the launched_date and end_date columns to datetime format

campaign_df['launch_date'] = pd.to_datetime(campaign_df['launch_date'])
campaign_df['end_date'] = pd.to_datetime(campaign_df['end_date'])

# Display the updated DataFrame to confirm the changes
print(campaign_df.dtypes)
```

```python
# Read the Excel file and create the crowdfunding_info_df
crowdfunding_info_df = pd.read_excel('Resources/crowdfunding.xlsx')

# Create and split the 'category & sub-category' column
campaign_df = crowdfunding_info_df.rename(columns={'category & sub-category': 'category'})
campaign_df[['category', 'subcategory']] = campaign_df['category'].str.split(' ', n=1, expand=True)

# Create category_df and subcategory_df without hard coding
category_df = campaign_df[['category']].drop_duplicates().reset_index(drop=True)
category_df['category_id'] = category_df.index + 1

subcategory_df = campaign_df[['subcategory']].drop_duplicates().reset_index(drop=True)
subcategory_df['subcategory_id'] = subcategory_df.index + 1

# Merge the campaign_df with the category_df on the "category" column and 
# the subcategory_df on the "subcategory" column.

merged_df = pd.merge(campaign_df, category_df, on='category')
merged_df = pd.merge(merged_df, subcategory_df, on='subcategory')


(merged_df.tail(10))
```
